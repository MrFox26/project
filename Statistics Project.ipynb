# MULTIPLE LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm
```

```python
x = datas[['GDP','GDP_growth','GDP_per_capita','GDP_per_capita_growth','Unemployment']]
y = datas['Inflation']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state =69)

model = LinearRegression()
model.fit(x_train,y_train)

x_train_with_const = sm.add_constant(x_train)
ols_model = sm.OLS(y_train,x_train_with_const)
ols_result = ols_model.fit()
print(ols_result.summary())
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm

# Assuming 'datas' is a pandas DataFrame with the required columns

# Function to perform multivariate regression for each country
def multivariate_regression_by_country(df, predictors, target):
    # Grouping the data by 'Country'
    results = {}
    
    for country, data in df.groupby('Country'):
        print(f"\nRegression for {country}:")
        
        # Select the predictor and target variables
        x = data[predictors]
        y = data[target]
        
        # Adding constant term for the intercept
        x_with_const = sm.add_constant(x)

        # Fit the OLS regression model using statsmodels
        ols_model = sm.OLS(y, x_with_const)
        ols_result = ols_model.fit()

        # Store the result in a dictionary
        results[country] = ols_result

        # Print the summary for each country
        print(ols_result.summary())
    
    return results

# List of predictor variables
predictors = ['GDP', 'GDP_growth', 'GDP_per_capita', 'GDP_per_capita_growth', 'Unemployment']

# Target variable (what we want to predict)
target = 'Inflation'

# Assuming 'datas' DataFrame has a 'Country' column along with the predictors and target
results = multivariate_regression_by_country(datas, predictors, target)
```
